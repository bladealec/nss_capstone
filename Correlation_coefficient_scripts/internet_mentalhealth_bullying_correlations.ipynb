## Finding Correlation Coefficient between USA child internet access and mental health / bullying

```python
import pandas as pd
```

```python
internet = pd.read_csv(r'C:\Users\alecb\Documents\NSS\Python\projects\alec_capstone\cleaned_data\USA_internet_access_everyone_and_3_to_17_clean2.csv')
mental = pd.read_csv(r'C:\Users\alecb\Documents\NSS\Python\projects\alec_capstone\cleaned_data\USA_child_suicide.csv')
```

```python
internet.head()
```

```python
internet['Percent of children with access'] = internet['Percent of children with access'].str.strip('%')
internet['Percent of children with access'] = internet['Percent of children with access'].astype(float) / 100
```

```python
internet.head()
```

```python
mental.head()
```

```python
print(internet.Year.unique())
print(mental.year.unique())
```

#### converting years for better merging

```python
for index, row in mental.iterrows():
    
    if row.year == 1989:
        mental.loc[index, 'year'] = 1991
        
    elif (row.year == 1998) | (row.year == 2000):
        mental.loc[index, 'year'] = 1999
        
    elif row.year == 2012:
        mental.loc[index, 'year'] = 2013
```

```python
internet_mental = pd.merge(internet, mental, 
                               left_on = 'Year', right_on = 'year', 
                               how = 'inner')
internet_mental.head()
```

```python
internet_mental.corr()
# planned suicide: -0.373, attempted suicide: -0.184
#fights: -0.817, sad or hopeless: 0.649
```
