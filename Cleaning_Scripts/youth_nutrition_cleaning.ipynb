#### checking to see if there is any difference between start and end years

```python
obesity['year_diff'] = obesity['YearStart'] - obesity['YearEnd']
#no differences
```

```python
obesity['year_diff'].unique()
```

#### reducing obesity df to only the columns I need / may need

```python
obesity = obesity[['YearStart', 'LocationDesc', 'Question', 'Data_Value', 'Data_Value_Alt',
         'Data_Value_Footnote', 'Sample_Size', 'Total', 'GeoLocation', 'StratificationID1']]
obesity
```

```python
len(obesity['LocationDesc'].unique())
#52 locations still
```

#### how many entries do not have data because of insufficient sample size

```python
obesity['Data_Value_Footnote'].unique()
```

```python
len(obesity[obesity['Data_Value_Footnote'] == 'Data not available because sample size is insufficient or data not reported.' ])
#2890 out of 11228 or 26%
```

#### if we only look at totals does that percentage decrease

```python
obesity_totals = obesity[obesity['StratificationID1'] == 'OVERALL']
obesity_totals
```

```python
len(obesity_totals[obesity_totals['Data_Value_Footnote'] == 'Data not available because sample size is insufficient or data not reported.' ])
#yes, all entries have sufficient sample sizes at the 'total' level
```

```python
obesity_totals = obesity_totals.drop(columns = ['Data_Value_Footnote'])
```

```python
obesity_totals.head()
```

#### Dropping and renaming columns

```python
obesity_totals = obesity_totals.drop(columns = ['Data_Value_Alt'])
```

```python
obesity_totals.rename(columns = {'YearStart': 'Year', 'LocationDesc': 'Location',
                                'Data_Value': 'Percentage_Value'})
```

```python
obesity_totals.to_csv('cleaned_data/USA_child_obesity_2001_2019.csv')
```
