### I don't like all the 'NaN' and would like to replace them with zeros, but I cant replace all 'NaN' becuase some occur after years where that country did have a high percentage of internet users. I need to find at what year did the first percentage of internet users appear. Then I will change the NaNs to 0 for all years before that.

```python
#running a test
```

```python
test = global_internet_use
```

```python
first_year = None
for col in test.columns:
    if test[col].max() > 0:
        first_year = col
        break

print(f"The first year with a number greater than 0 is {first_year}.")
```

```python
test['1989'][test['1989'] > 0]
```

```python
test['1990'][test['1990'] > 0]
```

```python
test[test.columns[0:32]] = test[test.columns[0:32]].fillna(0)
test
```

```python
#applying to actual df
```

```python
global_internet_use[global_internet_use.columns[1:32]] = global_internet_use[global_internet_use.columns[1:32]].fillna(0)
global_internet_use
```

### Now I would like to replace the rest of the NaN entries. If an entry is NaN, replace it with the previous year's percentage.

```python
#conducting a test
```

```python
for i, row in test.iterrows():
    for j in range(1, len(test.columns)):
        # Check if the current entry is null
        if pd.isnull(row[j]):
            # Replace the null with the previous year's value
            test.loc[i, test.columns[j]] = test.loc[i, test.columns[j-1]]
```

```python
test.info()
```

```python
#applying to actual df
```

```python
for i, row in global_internet_use.iterrows():
    for j in range(1, len(global_internet_use.columns)):
        # Checking if the current entry is null
        if pd.isnull(row[j]):
            # Replacing the null with the previous year's value
            global_internet_use.loc[i, global_internet_use.columns[j]] = global_internet_use.loc[i, global_internet_use.columns[j-1]]
```
