#### joining a table containing the countries populations throughout time

```python
populations = pd.read_csv('data/Globe Population/world_pop_by_country_1960_2020.csv')
populations.head()
```

#### I need to change the year columns into just one column containing all years, and then have a column for the population of that year

```python
populations = populations.drop(columns= ['Country Code', 'Indicator Name', 'Indicator Code'])
populations
```

```python
pop = populations.melt(
    id_vars=['Country Name'], 
    var_name='Year', 
    value_name='Population'
)
```

```python
pop
```

```python
pop.info()
```

```python
obesity_w_pop = pd.merge(global_obesity, pop, 
                               left_on = ('Location', 'Period'), right_on = ('Country Name', 'Year'), 
                               how = 'left')
```

```python
obesity_w_pop
```

```python
obesity_w_pop.info()
```

```python
obesity_w_pop = obesity_w_pop.drop_duplicates()
```

```python
obesity_w_pop.to_csv('cleaned_data/global_adults_obesity_1975_2016.csv')
```
