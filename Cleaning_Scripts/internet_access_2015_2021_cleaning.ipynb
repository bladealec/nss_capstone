```python
master_inter_access = pd.concat([internet_access_2015df, internet_access_2016df, internet_access_2017df,
                               internet_access_2018df, internet_access_2019df, internet_access_2020df,
                               internet_access_2021df], ignore_index=True)
master_inter_access
```

```python
master_inter_access['individuals_access'] = master_inter_access['individuals_access'].str.replace(',','')
master_inter_access['total_estimat'] = master_inter_access['total_estimat'].str.replace(',', '')
```

```python
master_inter_access
```

```python
master_inter_access['individuals_access'] = master_inter_access['individuals_access'].astype(int)
master_inter_access['total_estimat'] = master_inter_access['total_estimat'].astype(int)

master_inter_access['percent_of_children'] = master_inter_access['individuals_access'] / master_inter_access['total_estimat']
master_inter_access
```

```python
df_1989_2012
```

```python
df = master_inter_access.groupby('year', as_index = False).agg({'individuals_access':'sum', 'percent_of_children':'mean'})
```

```python
df['Country'] = 'United States of America'
df['Inividuals 3 - 17 (thousands)'] = df['individuals_access'] / 1000
df['Inividuals 3 - 17 (thousands)'] = df['Inividuals 3 - 17 (thousands)'].astype(int)
df['Description'] = 'has access to internet at home'

df
```
