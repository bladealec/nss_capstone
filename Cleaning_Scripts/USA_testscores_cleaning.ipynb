```python
import pandas as pd
```

```python
scores = pd.read_csv('data/Test Scores/state_grade_composite_scores.csv')
scores
```

```python
scores_1980_2000 = pd.read_csv('data/Test Scores/state_grade_composite_scores_1980_2000.csv')
scores_1980_2000
```

```python
scores_1980_2014 = pd.concat([scores_1980_2000, scores], ignore_index = True)
scores_1980_2014
```

```python
scores_1980_2014 = scores_1980_2014.drop(columns = ['Unnamed: 0'])
```

```python
scores_1980_2014.to_csv('data/Test Scores/master_stategrade_compositescores_1980_2014.csv')
```

```python
scores_1980_2014['year'].unique()
```

## Seperating scores by grade

```python
grade4 = scores_1980_2014[scores_1980_2014['grade'] == 4]
grade8 = scores_1980_2014[scores_1980_2014['grade'] == 8]
grade12 = scores_1980_2014[scores_1980_2014['grade'] == 12]
```
