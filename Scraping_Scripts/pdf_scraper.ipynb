## This script is used to scrape information from a pdf. The pdf's to be scraped are from the United States Census Bureau, and contain information about internet use from 1984 to 1997.

```python
import PyPDF2
import csv
```

```python
pdfFileObj = open('data/1997_internet_use_by_age.pdf', 'rb')

Reader = PyPDF2.PdfReader(pdfFileObj)

print(len(Reader.pages))

pageObj = Reader.pages[0]

print(pageObj.extract_text())

pdfFileObj.close()
```

```python
p = pageObj.extract_text()
first_list = p.splitlines()
first_list
```

```python
i=0
new_list=[]
while i<len(first_list):
  new_list.append(first_list[i:i+1])
  i+=1
```

```python
new_list
```

```python
new_list = [[i.split('|') for i in x] for x in new_list]
```

```python
new_list
```

```python
flattened_list  = [val for sublist in new_list for val in sublist]
```

```python
flattened_list
```

```python
with open('1997_internet_use_by_age.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(flattened_list)
```
