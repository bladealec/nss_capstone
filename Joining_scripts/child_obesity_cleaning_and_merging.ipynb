#### merging with global internet use

```python
internet = pd.read_csv(r'C:\Users\alecb\Documents\NSS\Python\projects\alec_capstone\cleaned_data\global_internet_use_1960_2022_clean.csv')
```

```python
internet.head()
```

```python
internet = internet.melt(
    id_vars=['Country Name'], 
    var_name='Year', 
    value_name='Percent Internet Use'
)

internet.head()
```

```python
print(obesity.info())
print(internet.info())
```

```python
obesity['Year'] = obesity['Year'].astype(str)
```

```python
internet['Country Name'] = internet['Country Name'].str.replace('United States', 'United States of America')
```

```python
globe_internet_obesity = pd.merge(obesity, internet, 
                               left_on = ('Year', 'Country'), right_on = ('Year', 'Country Name'), 
                               how = 'inner')

globe_internet_obesity.head()
```

```python
globe_internet_obesity = globe_internet_obesity.drop(columns = ['Country Name'])
globe_internet_obesity.head()
```

```python
globe_internet_obesity.to_csv(r'C:\Users\alecb\Documents\NSS\Python\projects\alec_capstone\cleaned_data\globe_internet_obesity_v2.csv')
```
